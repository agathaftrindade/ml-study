```python

class ZeroFilter(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
        
    def fit(self, X, y=None):
        return self
        
    def transform(self, X, y=None):
        X = X[X.humidity > 0] # Seems to be an outlier
        return X

class ComputedFeaturesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, gen_dryness=True):
        self.gen_dryness = gen_dryness 
        
    def fit(self, X, y=None):
        return self
        
    def transform(self, X, y=None):
        X = X.copy()
        if self.gen_dryness:
            X['is_dry'] = X['summary'].str.decode('utf-8').str.lower().str.contains('dry').map(lambda x: 1 if x else 0)
        return X

transform_pipeline = Pipeline([
    ('zero_filter', ZeroFilter()),
    ('comp_feature_adder', ComputedFeaturesAdder())
])


df = transform_pipeline.fit_transform(df)
pd.concat([
    df[df.is_dry == 0].sample(3),
    df[df.is_dry == 1].sample(3)
])[['summary', 'is_dry']]
```

## Task: Predict Apparent temperature based on temperature, humidity, wind_speed and maybe dryness

```python
def get_features_and_labels(df):
    X, Y = df[['temperature_c', 'humidity', 'wind_speed_kmh', 'is_dry']], df['apparent_temperature_c']
    return X, Y
```

```python
transform_pipeline = Pipeline([
    ('zero_filter', ZeroFilter()),
    ('comp_feature_adder', ComputedFeaturesAdder())
])

parsed_train_df = transform_pipeline.fit_transform(weather_train)
X_train, Y_train = get_features_and_labels(parsed_train_df)

X_train, Y_train
```

```python
transform_pipeline = Pipeline([
    ('zero_filter', ZeroFilter()),
    ('comp_feature_adder', ComputedFeaturesAdder())
])

parsed_validate_df = transform_pipeline.fit_transform(weather_validate)
X_validate, Y_validate = get_features_and_labels(parsed_validate_df)

X_validate, Y_validate
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import mean_squared_error
```

```python
def rmse(Y, predictions):
    mse = mean_squared_error(Y, predictions)
    return np.sqrt(mse)
```

### Base Estimator (prediction = temperature_c)

```python
class SimplePredictor(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, Y=None):
        pass

    def predict(self, X):
        return X.temperature_c
```

```python
simple_predictor = SimplePredictor()

simple_train_predictions = simple_predictor.predict(X_train)
simple_validation_predictions = simple_predictor.predict(X_validate)

print('simple_predictor train rmse: ', rmse(Y_train, simple_train_predictions))
print('simple_predictor validation rmse: ', rmse(Y_validate, simple_validation_predictions))
```
